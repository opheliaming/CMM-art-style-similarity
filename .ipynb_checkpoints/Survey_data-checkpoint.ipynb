```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('data/CCM Art Style Similarity_April 30, 2022_16.28.csv')
```

```python
mapping = pd.read_csv('data/Qualtrics_mapping.csv')
```

```python
data = data[(data['Status'] != 'Survey Preview') & (data['Finished'] == 'True')]   # filter only true responses
```

```python
data.shape
```

```python
#list(data.columns)
```

```python
# remove unwanted columns
unwanted = ['StartDate','EndDate','Status','IPAddress','Progress','Finished','RecordedDate','ResponseId',
            'RecipientLastName','RecipientFirstName','RecipientEmail','ExternalReference','LocationLatitude','LocationLongitude',
            'DistributionChannel','UserLanguage']
```

```python
 data = data[data.columns[~data.columns.isin(unwanted)]]
```

```python
# convert text to number
data = data.replace({'1 - Not at all': 1, '7 - Very similar': 7})
```

```python
#data.dtypes
```

```python
#data.head()
```

```python
open_end_question = ['1_visual_expl_1','1_style_expl_1','1_visual_expl_2','1_style_expl_2','art_training_type']
```

```python
numeric_columns = list(set(data.columns).difference(open_end_question))
```

```python
data[numeric_columns] = data[numeric_columns].astype('int32')
```

```python
# demographic 
sns.distplot(data.age)
```

```python
data.describe()
```

```python
DF_score = pd.DataFrame({'question': data.iloc[:, 1:145].columns,
                         'mean': data.iloc[:, 1:145].mean(axis=0)})
DF_score = DF_score.reset_index(drop = True)
DF_score['question_number'] = DF_score['question'].apply(lambda x: int(x.split("_")[0]))
DF_score['question_type'] = DF_score['question'].apply(lambda x: x.split("_")[2])
```

```python
DF_score_pivot = pd.pivot_table(DF_score, values='mean', index= 'question_number',
                    columns=['question_type'])
DF_score_pivot = DF_score_pivot.reset_index()
DF_score_pivot.columns = ['pair_id', 'general_score','style_score','subject_score']
```

```python
DF_merge = pd.merge(DF_score_pivot, mapping, left_on = 'pair_id', right_on = 'Pair_ID')
DF_merge = DF_merge[['pair_id','QualID_A','File_A','QualID_B','File_B',
                     'general_score','style_score','subject_score','Style_match','Subject_match']]
```

```python
# Overall statistics
DF_merge[['general_score','style_score','subject_score']].describe()
```

```python
# Group by style match 
round(DF_merge[['general_score','style_score','subject_score','Style_match']].groupby('Style_match').mean().reset_index(),2)
```

```python
# Group by subject match 
round(DF_merge[['general_score','style_score','subject_score','Subject_match']].groupby('Subject_match').mean().reset_index(),2)
```

```python
# Group by both style and subject match 
round(DF_merge[['general_score','style_score','subject_score','Style_match','Subject_match']].groupby(['Style_match','Subject_match']).mean().reset_index(),2)
```

```python
# correlation of general_score, style_score, subject_score
corr = round(DF_merge[['general_score','style_score','subject_score']].corr(),2)
corr.style.background_gradient(cmap='coolwarm')
```
