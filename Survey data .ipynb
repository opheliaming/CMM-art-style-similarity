```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('data/CCM Art Style Similarity_April 28, 2022_14.59.csv')
```

```python
data = data[data['Status'] == '0']   # filter only true responses
```

```python
# remove unwanted columns
unwanted = ['StartDate','EndDate','Status','IPAddress','Progress','Finished','RecordedDate','ResponseId',
            'RecipientLastName','RecipientFirstName','RecipientEmail','ExternalReference','LocationLatitude','LocationLongitude',
            'DistributionChannel','UserLanguage']
```

```python
 data = data[data.columns[~data.columns.isin(unwanted)]]
```

```python
#list(data.columns)
```

```python
open_end_question = ['1_visual_expl_1','1_style_expl_1','1_visual_expl_2','1_style_expl_2','art_training_type']
```

```python
numeric_columns = list(set(data.columns).difference(open_end_question))
```

```python
data[numeric_columns] = data[numeric_columns].astype('int32')
```
